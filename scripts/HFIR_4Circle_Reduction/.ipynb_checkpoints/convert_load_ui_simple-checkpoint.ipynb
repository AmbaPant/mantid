```python
import os
```

```python
files = os.listdir(os.path.join(os.getcwd(), 'pyvdrive/interface'))
py_files = [file_name for file_name in files if file_name.endswith('.py')]
```

```python
dir_name = os.path.join(os.getcwd(), 'pyvdrive/interface')
for py_name in py_files:
    # load file
    py_file = open(os.path.join(dir_name, py_name), 'r')
    py_lines = py_file.readlines()
    py_file.close()
    
    # find line with gui.ui_
    ui_name = None
    is_simple_file = False
    for line_number, line in enumerate(py_lines):
        if line.count('gui.ui_'):
            # print line_number, ' : ', line
            is_simple_file = True
            ui_name = line.split('gui.ui_')[0].split()[0] + '.ui'
            break
    # END-FOR
    
    # skip those without load_ui
    if is_simple_file:
        print py_name, ' : ', ui_name
    else:
        continue
        
    # locate those lines
    import_ui_line_number = None
    pyqt5_line_number = None
    pyqt4_line_number = None
    load_ui_line_number = None
    set_ui_line = None
    init_ui_line = None
    
    # check whether there is a 'load_ui' inside and 'from PyQt5' and 'from PyQt4'
    for line_number, line in enumerate(py_lines):
        if line.count('gui.ui_') and line.count('import'):
            print line_number, ' : ', line
            import_ui_line_number = line_number
        elif line.count('from PyQt5') and pyqt5_line_number is None:
            print line_number, ' : ', line
            pyqt5_line_number = line_number
        elif line.count('from PyQt4') and pyqt4_line_number is None:
            print line_number, ' : ', line
            pyqt4_line_number = line_number
        elif line.count('self.ui = '):
            print line_number, ' : ', line
            init_ui_line = line_number
        elif line.count('ui.setupUi'):
            print line_number, ' : ', line
            set_ui_line = line_number

    # END-FOR
    
        
    # replace
    # Qt file line
    line_number = pyqt5_line_number
    print line_number, ' : '
    print '- '
    print py_lines[line_number]
    py_lines[line_number] += '    from PyQt5.QtWidgets import QVBoxLayout\n    from PyQt5.uic import loadUi as load_ui\n'
    print '+'
    print py_lines[line_number]
    print
    
    # Qt4 line
    line_number = pyqt4_line_number
    print line_number, ' : \n-'
    print py_lines[line_number]
    py_lines[line_number] += '    from PyQt4.QtGui import QVBoxLayout\n    from PyQt4.uic import loadUi as load_ui\n'
    print '+'
    print py_lines[line_number]
    print
    
    # import line
    level1 = '        '
    line_number = import_ui_line_number
    print line_number, ' : \n-'
    print py_lines[line_number]
    py_lines[line_number] = '\n'
    print '+'
    print py_lines[line_number]
    print
    
    # load_ui line
    level1 = '        '
    line_number = init_ui_line
    print line_number, ' : \n-'
    print py_lines[line_number]
    py_lines[line_number] = level1 + 'ui_path = os.path.join(os.path.dirname(__file__), "gui/{}")\n'.format(ui_name)
    py_lines[line_number] += level1 + 'self.ui = load_ui(ui_path, baseinstance=self)\n'
    print '+'
    print py_lines[line_number]
    print
    
    # delete line to setupUI
    line_number = set_ui_line
    print line_number, ' : \n-'
    print py_lines[line_number]
    py_lines[line_number] = ''
    print '+'
    print py_lines[line_number]
    print
    
    # write
    wbuf = ''
    for line in py_lines:
        wbuf += line
    
    out_file = open(os.path.join(dir_name, py_name), 'w')
    out_file.write(wbuf)
    out_file.close()

    
# END-FOR
    
```
